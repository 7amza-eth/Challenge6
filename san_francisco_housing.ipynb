# Housing Rental Analysis for San Francisco

```python
# Import the required libraries and dependencies
import pandas as pd
import hvplot.pandas
from pathlib import Path
```

## Import the data 

```python
#Create dataframe from csv and review
sfo_data_df = pd.read_csv(
    Path('Resources/sfo_neighborhoods_census_data.csv')
)
display(sfo_data_df.head())
display(sfo_data_df.tail())
```

---

## Calculate and Plot the Housing Units per Year

```python
# Group data by year then average
housing_units_by_year = sfo_data_df.groupby('year').mean()
housing_units_by_year
```

```python
# Plot number of housing units by year
housing_units_by_year.hvplot.bar(
    x='year',
    y='housing_units',
    title='Housing Units By Year',
    ylabel='Housing Units',
    xlabel='Year',
    ylim=(370000,390000),
    yformatter='%f'
)
```

**Question:** What is the overall trend in housing_units over the period being analyzed?

**Answer:**   There is a overall uptrend in the number of housing units between 2010 to 2016

---

## Calculate and Plot the Average Sale Prices per Square Foot

```python
# Group data by year and average
prices_square_foot_by_year = sfo_data_df.groupby('year').mean()
prices_square_foot_by_year
```

**Question:** What is the lowest gross rent reported for the years included in the DataFrame?

**Answer:** The lowest gross rent was $1239 which occured in 2010

```python
# Remove housing_units column
prices_square_foot_by_year = prices_square_foot_by_year.drop(columns='housing_units')
prices_square_foot_by_year
```

```python
# Plot price per square foot and gross rent by year
by_year_plot = prices_square_foot_by_year.hvplot(
    x='year'
).opts(
    title='Price per Square Foot vs Gross Rent by Year', 
    xlabel='Year',
    ylabel='Price', 
    width=1000, 
    height=500,
    legend_position='top_left',
)
by_year_plot
```

**Question:** Did any year experience a drop in the average sale price per square foot compared to the previous year?

**Answer:** There was a drop in average sale price per square foot in 2011 when compared to the price in 2010.

**Question:** If so, did the gross rent increase or decrease during that year?

**Answer:** Gross rent still increased during that year.

## Compare the Average Sale Prices by Neighborhood

```python
# Group by year and neighborhood then average
prices_by_year_by_neighborhood = sfo_data_df.groupby(['year', 'neighborhood']).mean()
prices_by_year_by_neighborhood
```

```python
# Removing housing_units
prices_by_year_by_neighborhood = prices_by_year_by_neighborhood.drop(columns='housing_units')
display(prices_by_year_by_neighborhood.head())
display(prices_by_year_by_neighborhood.tail())
```

```python
# Plot Price per square foot and gross rent by year with each neighborhood on a seperate graph
neighborhood_sqr_foot_plot = prices_by_year_by_neighborhood.hvplot(
    x='year',
    label='Price/SqrFt',
    groupby='neighborhood'
).opts(
    title="Price per Square Foot vs Gross Rent by Year for each Neighborhood", 
    ylabel="Price", 
    xlabel='Year',
    width=1000, 
    height=500,
    legend_position='top_left'
)
neighborhood_sqr_foot_plot
```

**Question:** For the Anza Vista neighborhood, is the average sale price per square foot for 2016 more or less than the price thatâ€™s listed for 2012? 

**Answer:** The avarage sale price/squar foot for 2016 is less than what it was in 2016 for the Anza Vista neighborhood

## Build an Interactive Neighborhood Map

```python
# Create datafram from csv
neighborhood_locations_df = pd.read_csv(
    Path('Resources/neighborhoods_coordinates.csv'),
    index_col='Neighborhood'
)
neighborhood_locations_df
```

```python
# Groupby neighborhood then average
all_neighborhood_info_df = sfo_data_df.groupby('neighborhood').mean()
all_neighborhood_info_df
```

```python
# Combine both dataframes by columns
all_neighborhoods_df = pd.concat(
    [neighborhood_locations_df, all_neighborhood_info_df], 
    axis="columns",
    sort=False
)
display(all_neighborhoods_df.head())
display(all_neighborhoods_df.tail())
```

```python
# Clean up dataframe
all_neighborhoods_df = all_neighborhoods_df.reset_index().dropna()
# Rename the "index" column as "Neighborhood" for use in the Visualization
all_neighborhoods_df = all_neighborhoods_df.rename(columns={"index": "Neighborhood"})
display(all_neighborhoods_df.head())
display(all_neighborhoods_df.tail())
```

```python
# Plot neighborhood data on geograph
all_neighborhoods_df.hvplot.points(
    'Lon',
    'Lat',
    geo=True,
    size='sale_price_sqr_foot',
    color='gross_rent',
    tiles='OSM',
    xlabel='Longitude',
    ylabel='Latitude',
    frame_width=700,
    frame_height=500,
    title='Price/sqft (dot size) & Gross Rent (color) by Neighborhood',
    hover_cols= ['Neighborhood', 'sale_price_sqr_foot', 'gross_rent']
)
```

**Question:** Which neighborhood has the highest gross rent, and which has the highest sale price per square foot?

**Answer:** Westwood Park has the heighest gross rent, while Union Square District has the highest sale price per square foot.
